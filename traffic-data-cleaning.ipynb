## Subsetting the dataframe to get only Los Angeles County data. 

```python
la_data = df[df['County Name'] == 'Los Angeles County']
```

### Reformatting the Date Column to be ISO 8601 format.

```python
la_data['Date'] = la_data['Date'].str.replace('/', '-')
```

```python
la_data.head(2)
```

```python
la_data.info()
```

### Converting the Date column to datetime data type.

```python
la_data['Date'] = pd.to_datetime(la_data['Date'])
la_data.info()
```

```python
la_data.to_csv('datasets/traffic/la_data.csv', index=False)
```
