### LA data

```python
cali_counties = pd.read_csv('./datasets/Covid-19-Data/latimes-county-totals.csv')
cali_counties
```

```python
cali_counties.info()
```

```python
cali_counties.isna().sum()
```

```python
# get LA county
la_county = cali_counties.loc[cali_counties.county == 'Los Angeles']
la_county
```

```python
la_county['date'] = pd.to_datetime(la_county['date'])
la_county.set_index('date', inplace = True)
la_county.sort_index(inplace = True)
```

```python
la_county.head(2)
```

```python
la_county.info()
```

```python
# locate only 2020 rows
la_county_2020 = la_county.loc['2020']
la_county_2020
```

```python
# replace NaN with zeroes since on January 26 2020 there was only one covid-19
# case in LA on this date (first case), no death and no new cases
la_county_2020.replace({np.nan: 0}, inplace=True)
```

```python
la_county_2020.info()
```

```python
la_county_2020.head()
```

```python
# save df
la_county_2020.to_csv('./datasets/Covid-19-Data/LA_county_covid_2020.csv')
```

### NYC data

```python
nyc = pd.read_csv('./datasets/Covid-19-Data/NYC_COVID-19_Daily_Counts_of_Cases_Hospitalizations_and_Deaths.csv')
nyc
```

```python
# make date in correct ISO8601 format
for i in range(0, nyc.shape[0]):
    nyc['DATE_OF_INTEREST'].iloc[i] = '-'.join([nyc['DATE_OF_INTEREST'][i].split('/')[2], nyc['DATE_OF_INTEREST'][i].split('/')[0], nyc['DATE_OF_INTEREST'][i].split('/')[1]])
    
    
```

```python
nyc
```

```python
nyc.rename(columns={'DATE_OF_INTEREST': 'date'}, inplace=True)
nyc['date'] = pd.to_datetime(nyc['date'])
nyc.set_index('date', inplace = True)
nyc.sort_index(inplace = True)
```

```python
nyc.info()
```

```python
# locate only 2020 rows
# first case in NYC was on Feb 29 2020
nyc_boroughs_2020 = nyc.loc['2020']
nyc_boroughs_2020.info()
```

```python
nyc_boroughs_2020.head(2)
```

```python
nyc_boroughs_2020.tail(2)
```

```python
nyc_boroughs_2020.to_csv('./datasets/Covid-19-Data/NYC_all_boroughs_covid_2020.csv')
```
